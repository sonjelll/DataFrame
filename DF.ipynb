```python
import pandas as pd

df_excel = pd.read_excel('data.xlsx', sheet_name='data')
df_excel
```

```python
# 2 total pada tahun tertentu
tahun_tertentu = 2021
# Inisialisasi variabel untuk total produksi sampah
total_sampah_tahun_tertentu = 0

# Mengiterasi baris menggunakan iterrows()
for index, row in df_excel.iterrows():
    if row['tahun'] == tahun_tertentu:  # Filter berdasarkan tahun
        total_sampah_tahun_tertentu += row['jumlah_produksi_sampah']  # Jumlahkan produksi sampah

# Tampilkan hasil total sampah untuk tahun tertentu
print(f"Total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat pada tahun {tahun_tertentu}: {total_sampah_tahun_tertentu} ton")


# 3. total per tahun
total_per_tahun = {}

# Mengiterasi baris menggunakan iterrows()
for index, row in df_excel.iterrows():
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']
    
    # Jika tahun belum ada dalam dictionary, inisialisasi dengan 0
    if tahun not in total_per_tahun:
        total_per_tahun[tahun] = 0
    
    # Menambahkan produksi sampah per tahun
    total_per_tahun[tahun] += sampah

# Tampilkan hasil total per tahun
print("Total produksi sampah per tahun:")
for tahun, total in total_per_tahun.items():
    print(f"Tahun {tahun}: {total} ton")


# 4. total per Kota/Kabupaten per tahun
total_per_kabupaten_tahun = {}

# Mengiterasi baris menggunakan iterrows()
for index, row in df_excel.iterrows():
    kabupaten_kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']
    
    # Jika kombinasi kota dan tahun belum ada dalam dictionary, inisialisasi dengan 0
    if (kabupaten_kota, tahun) not in total_per_kabupaten_tahun:
        total_per_kabupaten_tahun[(kabupaten_kota, tahun)] = 0
    
    # Menambahkan produksi sampah per kota dan tahun
    total_per_kabupaten_tahun[(kabupaten_kota, tahun)] += sampah

# Tampilkan hasil total per Kota/Kabupaten per tahun
print("Total produksi sampah per Kota/Kabupaten per tahun:")
for (kabupaten_kota, tahun), total in total_per_kabupaten_tahun.items():
    print(f"{kabupaten_kota} - Tahun {tahun}: {total} ton")
```

```python
# Menyimpan hasil total per tahun ke CSV dan Excel
pd.DataFrame(list(total_per_tahun.items()), columns=['Tahun', 'Total_Produksi_Sampah']).to_csv('total_per_tahun.csv', index=False)
pd.DataFrame(list(total_per_tahun.items()), columns=['Tahun', 'Total_Produksi_Sampah']).to_excel('total_per_tahun.xlsx', index=False)

# Menyimpan hasil total per Kota/Kabupaten per tahun ke CSV dan Excel
pd.DataFrame(list(total_per_kabupaten_tahun.items()), columns=['Kabupaten_Kota_Tahun', 'Total_Produksi_Sampah']).to_csv('total_per_kabupaten_tahun.csv', index=False)
pd.DataFrame(list(total_per_kabupaten_tahun.items()), columns=['Kabupaten_Kota_Tahun', 'Total_Produksi_Sampah']).to_excel('total_per_kabupaten_tahun.xlsx', index=False)
```
